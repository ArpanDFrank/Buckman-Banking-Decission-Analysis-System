## Implementing ID3 Model

```python
import numpy as np
import math
```

```python
def calculate_entropy(data, target_column): # for each categorical variable
    total_rows = len(data)
    target_values = data[target_column].unique()
 
    entropy = 0
    for value in target_values:
        # Calculate the proportion of instances with the current value
        value_count = len(data[data[target_column] == value])
        proportion = value_count / total_rows
        entropy -= proportion * math.log2(proportion) if proportion != 0 else 0
 
    return entropy
```

```python
def calculate_information_gain(data, feature, target_column):
    entropy_outcome=calculate_entropy(data,target_column)
    # Calculate weighted average entropy for the feature
    unique_values = data[feature].unique()
    weighted_entropy = 0
 
    for value in unique_values:
        subset = data[data[feature] == value]
        proportion = len(subset) / len(data)
        weighted_entropy += proportion * calculate_entropy(subset, target_column)
 
    # Calculate information gain
    information_gain = entropy_outcome - weighted_entropy
 
    return information_gain
```

```python
def decision_tree(data, target_column, features):
    # Base case: if all target values are the same, return that value
    if len(data[target_column].unique()) == 1:
        return data[target_column].iloc[0]

    # Base case: if there are no features left, return the majority class
    if len(features) == 0:
        return data[target_column].mode()[0]

    # Select the feature that provides the highest information gain
    information_gains = {feature: calculate_information_gain(data, feature, target_column) for feature in features}
    best_feature = max(information_gains, key=information_gains.get)

    # Create a node with the best feature
    tree = {best_feature: {}}

    # Remove the best feature from the list of features
    features = [feature for feature in features if feature != best_feature]

    # Recursively build the subtree for each value of the best feature
    for value in data[best_feature].unique():
        subset = data[data[best_feature] == value]
        subtree = decision_tree(subset, target_column, features)
        tree[best_feature][value] = subtree
    return tree
```

```python
def predict(tree, data):
    predictions = []
    for index, row in data.iterrows():
        current_tree = tree
        while isinstance(current_tree, dict):
            feature = list(current_tree.keys())[0]  # Get the feature at the current level
            value = row[feature]  # Get the value of the feature from the current row
            current_tree = current_tree[feature].get(value)  # Traverse the tree
        if current_tree is not None:
            predictions.append(current_tree)
        else:
            predictions.append('Unknown')
    return predictions
```

```python

def plot_tree(tree, parent=None, depth=0):
    if isinstance(tree, dict):
        for key, value in tree.items():
            if parent is not None:
                print("|   " * depth, end="")
                print("|--- ", end="")
                print(parent)

            print("|   " * depth, end="")
            print("|--- ", end="")
            print(key)
            plot_tree(value, key, depth + 1)
    else:
        print("|   " * depth, end="")
        print("|--- ", end="")
        print(tree)
```

```python
def arrange_xtest_data(x_test, tree):
    tree_features = list(tree.keys())  # Get the features used in the decision tree
    
    # Arrange features from x_test according to the order of features in the decision tree
    arranged_features = [feature for feature in tree_features if feature in x_test.columns]
    remaining_features = [feature for feature in x_test.columns if feature not in tree_features]
    arranged_features.extend(remaining_features)
    
    # Reorder columns in x_test DataFrame
    x_test = x_test[arranged_features]
    
    return x_test
```
